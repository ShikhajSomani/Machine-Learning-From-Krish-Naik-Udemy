```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('Algerian_forest_fires_dataset_UPDATE.csv',header=1)
```

# Data Cleaning

```python
# missing values
df[df.isnull().any(axis=1)]
```

```python
df.loc[:122,"Region"] = 0
df.loc[122:,"Region"] = 1
```

```python
df.info()
```

```python
df[['Region']] = df[['Region']].astype(int)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
# Remove null values
df = df.dropna().reset_index(drop = True)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.iloc[[122]]
```

```python
# Removing 122th index
df = df.drop(122).reset_index(drop=True)
```

```python
df.iloc[[122]]
```

```python
df.columns
```

```python
# Fix the spaces in the columns names
df.columns = df.columns.str.strip()
df.columns
```

```python
df.info()
```

# Changes the required columns as int data type

```python
df.columns
```

```python
df[['day', 'month', 'year', 'Temperature', 'RH', 'Ws']] = df[['day', 'month', 'year', 'Temperature', 'RH', 'Ws']].astype(int)
```

```python
df.info()
```

# Changing the other column to float datatypes

```python
df[['Rain', 'FFMC',
       'DMC', 'DC', 'ISI', 'BUI', 'FWI']] = df[['Rain', 'FFMC',
       'DMC', 'DC', 'ISI', 'BUI', 'FWI']].astype(float)
```

```python
df.info()
```

```python
df.describe()
```

```python
df.head()
```

```python
# Lets save the cleaned dataset
df.to_csv('Algerian_forest_fires_cleaned_dataset.csv', index=False)
```

# Exploratory Data Analysis

```python
# Drop day,month,year in the dataset
df_copy = df.drop(['day','month','year'], axis=1)
```

```python
df_copy.head()
```

```python
# categories in classes
df_copy['Classes'].value_counts()
```

```python
# Encoding of the categories in classes
df_copy['Classes'] = np.where(df_copy['Classes'].str.contains('not fire'),0,1)
# The classes that have values "not fire" put the value as 0 or else 1
```

```python
df_copy.head()
```

```python
df_copy.tail()
```

```python
df_copy['Classes'].value_counts()
```

```python
# Plot density plot for all features
import seaborn as sns
sns.set()
df_copy.hist(bins=50,figsize=(20,15))
plt.show()
```

```python
# Percentage Pie Chart
percentage = df_copy['Classes'].value_counts(normalize = True)*100
```

```python
# Plotting Pie Chart
classlabels = ["Fire","Not Fire"]
plt.figure(figsize=(12,7))
plt.pie(percentage,labels=classlabels,autopct='%1.1f%%')
plt.title("Pie Chart of Classes")
plt.show()
```

```python
# Correlation
df_copy.corr()
```
